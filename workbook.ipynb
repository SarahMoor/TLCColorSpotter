```python
import cv2
import numpy as np
import scipy as sp
import scipy.signal
import math as m

import os, os.path

import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
def show(img):
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_RGB2BGR))
```

```python
nemo = cv2.imread("nemo.jpeg")
nemo_bw = cv2.imread("nemo.jpeg", cv2.IMREAD_GRAYSCALE)
```

```python
print(nemo.shape) # the shape of this image is 3-dimensional where the 3rd dimension holds each RGB value
show(nemo)
```

```python
print(nemo_bw.shape) # the shape of this image is 2-dimensional where because it is black/white
show(nemo_bw)
```
